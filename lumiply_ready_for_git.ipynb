```python
import os
import shutil

src_root = "/content/drive/MyDrive/LumiNet_Files/LumiNet-lee"  # 원본 폴더
dst_root = "/content/drive/MyDrive/LumiNet_Files/LumiNet-lee-small"  # 작은 파일만 모을 폴더

max_bytes = 100 * 1024 * 1024  # 100MB

os.makedirs(dst_root, exist_ok=True)

copied_files = []  # 복사된 파일 목록
skipped_files = [] # 용량 때문에 스킵된 파일 목록

for root, dirs, files in os.walk(src_root):
    # 상대 경로 계산
    rel_dir = os.path.relpath(root, src_root)
    target_dir = os.path.join(dst_root, rel_dir)
    os.makedirs(target_dir, exist_ok=True)

    for f in files:
        src_path = os.path.join(root, f)
        try:
            size = os.path.getsize(src_path)
        except OSError:
            print(f"[ERROR] 파일 정보 읽기 실패: {src_path}")
            continue

        size_mb = size / 1024 / 1024
        rel_path = os.path.relpath(src_path, src_root)  # 로그 깔끔하게 보이게 상대 경로로 표시

        # 여기서 필터링: max_bytes보다 작을 때만 복사
        if size <= max_bytes:
            dst_path = os.path.join(target_dir, f)
            shutil.copy2(src_path, dst_path)
            copied_files.append((rel_path, size_mb))
            print(f"[COPIED] {rel_path} ({size_mb:.2f} MB)")
        else:
            skipped_files.append((rel_path, size_mb))
            print(f"[SKIP  ] {rel_path} ({size_mb:.2f} MB)  ### 용량 초과")

# ==========================
# 요약 로그 출력
# ==========================
print("\n================ SUMMARY ================")
print(f"총 복사된 파일 수 : {len(copied_files)}")
print(f"총 스킵된 파일 수 : {len(skipped_files)} (max {max_bytes/1024/1024:.0f} MB 초과)")

if skipped_files:
    print("\n[용량 때문에 스킵된 파일 목록]")
    for rel_path, size_mb in skipped_files:
        print(f" - {rel_path}  ({size_mb:.2f} MB)")
```
